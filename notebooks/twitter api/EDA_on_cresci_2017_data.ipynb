```python
trace = go.Histogram(x=real_users.followers_count, xbins=dict(start=np.min(real_users.followers_count), 
                                                              size=20, end=real_users.followers_count.quantile(0.8)),
                     opacity=0.75, histnorm='probability', name='genuine')
trace2 = go.Histogram(x=spam_users.followers_count, xbins=dict(start=np.min(real_users.followers_count), 
                                                               size=20, end=real_users.followers_count.quantile(0.8)), 
                      histnorm='probability', opacity=0.75, name='spam')
layout = dict(title='Follower Count Distribution', xaxis=dict(title='Number of Followers'), yaxis=dict(title='Proportion'), barmode='overlay')
fig = dict(data=[trace, trace2], layout=layout)
plotly.offline.iplot(fig)
```

### EDA on Tweet Data

```python
real_tweets = real_tweets.fillna(0)
spam_tweets = spam_tweets.fillna(0)
```

```python
# Capture whether the tweet contains retweet
real_tweets['contain_RT'] = pd.to_numeric(real_tweets['text'].str.contains('RT'))
spam_tweets['contain_RT'] = pd.to_numeric(spam_tweets['text'].str.contains('RT'))
```

```python
# Capture tweet length
real_tweets['tweet_len'] = real_tweets['text'].str.len()
spam_tweets['tweet_len'] = spam_tweets['text'].str.len()
```

```python
real_tweets_mean = real_tweets.groupby(by='user_id').mean()
spam_tweets_mean = spam_tweets.groupby(by='user_id').mean()
```

```python
spam_tweets_mean.describe()
```

```python
real_tweets_mean.describe()
```
